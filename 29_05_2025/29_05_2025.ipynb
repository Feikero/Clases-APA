```python
import tkinter as tk
from tkinter import ttk

def muestra_mascota(label, imagen):
    def muestra():
        label.configure(image=imagen)
        label.image = imagen  # También se puede hacer esto para evitar que se pierda
    return muestra

def escoge_mascota(raiz, mascotas, var_mascota):
    marco_sel = ttk.LabelFrame(raiz, text='Escoge tu mascota favorita:', labelanchor=tk.N)
    marco_sel.pack(side=tk.LEFT, anchor=tk.N)

    marco_vis = ttk.Frame(raiz)
    marco_vis.pack(side=tk.LEFT)

    label_vis = ttk.Label(marco_vis)
    label_vis.pack()

    imagenes = []  # ← Lista para evitar que el recolector de basura elimine las imágenes

    for value, mascota in enumerate(mascotas):
        imagen = tk.PhotoImage(file=mascota[1])
        imagenes.append(imagen)  # Guardamos la imagen
        radio = ttk.Radiobutton(
            marco_sel,
            text=mascota[0],
            value=value,
            variable=var_mascota,
            command=muestra_mascota(label_vis, imagen)
        )
        radio.pack()

    terminar = ttk.Button(marco_sel, text='terminar', command=raiz.destroy)
    terminar.pack(side=tk.BOTTOM)

    # Guardamos las imágenes en la raíz para que no se pierdan
    raiz.imagenes = imagenes

def main():
    win = tk.Tk()
    ttk.Style().configure('.', font=('Arial', 16), background='cornflower blue', padding=8)
    mascotas = (
        ('Lindo gatito 1', 'gatito1.png'),
        ('Lindo gatito 2', 'gatito2.png'),
        ('Lindo gatito 3', 'gatito3.png'),
        ('Rata inmunda', 'rata.png'),
    )

    var_mascota = tk.IntVar(value=-1)
    escoge_mascota(win, mascotas, var_mascota)
    win.mainloop()
    return var_mascota

if __name__ == '__main__':
    var_mascota = main()
    print(var_mascota.get())
```

```python
import tkinter as tk
from tkinter import ttk

def volum (arrel, varVolum):
    marcVolum = ttk.LabelFrame(arrel, text='Volum', padding = 8)
    marcVolum.pack()
    volum = ttk.Scale(marcVolum, from_ = 0, to = 1, length = 200, value = varVolum.get(), variable = varVolum)
    volum.pack()

def main():
    win = tk.Tk()
    ttk.Style().configure('.', font =('Arial', 16), padding =8)
    varVolum = tk.DoubleVar(value = 0.6)
    volum(win, varVolum)
    win.mainloop()

    return varVolum

if __name__ == '__main__':
    varVolum = main()
    print(varVolum.get())
```
